```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import genfromtxt
import itertools
```

## 1. Plot Training Loss History

```python
Model1_Fidelity, Model2_Fidelity, Model1_HSinner, Model2_HSinner = [], [], [], []

for i in range(5):
    Model1_Fidelity_PATH = f'/Users/tak/Github/QEmbedding/Results/earlystop 10 experiments/experiment{i+1}/Model1 Fidelity/Training_Loss_History.txt'
    Model1_HSinner_PATH = f'/Users/tak/Github/QEmbedding/Results/earlystop 10 experiments/experiment{i+1}/Model1 HSinner/Training_Loss_History.txt'
    Model2_Fidelity_PATH = f'/Users/tak/Github/QEmbedding/Results/earlystop 10 experiments/experiment{i+1}/Model2 Fidelity/Training_Loss_History.txt'
    Model2_HSinner_PATH = f'/Users/tak/Github/QEmbedding/Results/earlystop 10 experiments/experiment{i+1}/Model2 HSinner/Training_Loss_History.txt'

    Model1_Fidelity.append(genfromtxt(Model1_Fidelity_PATH, delimiter=',').tolist())
    Model1_HSinner.append(genfromtxt(Model1_HSinner_PATH, delimiter=',').tolist())
    Model2_Fidelity.append(genfromtxt(Model2_Fidelity_PATH, delimiter=',').tolist())
    Model2_HSinner.append(genfromtxt(Model2_HSinner_PATH, delimiter=',').tolist())

Model1_Fidelity = list(itertools.zip_longest(*Model1_Fidelity, fillvalue=np.nan))
Model1_HSinner = list(itertools.zip_longest(*Model1_HSinner, fillvalue=np.nan))
Model2_Fidelity = list(itertools.zip_longest(*Model2_Fidelity, fillvalue=np.nan))
Model2_HSinner = list(itertools.zip_longest(*Model2_HSinner, fillvalue=np.nan))

Model1_Fidelity_mean = np.array([np.nanmean(i) for i in Model1_Fidelity])
Model1_Fidelity_std = np.array([np.nanstd(i) for i in Model1_Fidelity])

Model1_HSinner_mean = np.array([np.nanmean(i) for i in Model1_HSinner])
Model1_HSinner_std = np.array([np.nanstd(i) for i in Model1_HSinner])

Model2_Fidelity_mean = np.array([np.nanmean(i) for i in Model2_Fidelity])
Model2_Fidelity_std = np.array([np.nanstd(i) for i in Model2_Fidelity])

Model2_HSinner_mean = np.array([np.nanmean(i) for i in Model2_HSinner])
Model2_HSinner_std = np.array([np.nanstd(i) for i in Model2_HSinner])
```

```python
plt.rcParams['figure.figsize'] = [15, 5]
fig, (ax1, ax2) = plt.subplots(1,2)
clrs = sns.color_palette("husl", 4)
with sns.axes_style("darkgrid"):
    ax1.plot(range(len(Model1_Fidelity_mean)), Model1_Fidelity_mean, label="Model1_Fidelity", c=clrs[0])
    ax1.fill_between(range(len(Model1_Fidelity_mean)), Model1_Fidelity_mean-Model1_Fidelity_std, Model1_Fidelity_mean+Model1_Fidelity_std, alpha=0.3,facecolor=clrs[0])

    ax1.plot(range(len(Model1_HSinner_mean)), Model1_HSinner_mean, label="Model1_HSinner", c=clrs[1])
    ax1.fill_between(range(len(Model1_HSinner_mean)), Model1_HSinner_mean-Model1_HSinner_std, Model1_HSinner_mean+Model1_HSinner_std, alpha=0.3,facecolor=clrs[1])

    ax2.plot(range(len(Model2_Fidelity_mean)), Model2_Fidelity_mean, label="Model2_Fidelity", c=clrs[2])
    ax2.fill_between(range(len(Model2_Fidelity_mean)), Model2_Fidelity_mean-Model2_Fidelity_std, Model2_Fidelity_mean+Model2_Fidelity_std, alpha=0.3,facecolor=clrs[2])

    ax2.plot(range(len(Model2_HSinner_mean)), Model2_HSinner_mean, label="Model2_HSinner", c=clrs[3])
    ax2.fill_between(range(len(Model2_HSinner_mean)), Model2_HSinner_mean-Model2_HSinner_std, Model2_HSinner_mean+Model2_HSinner_std, alpha=0.3,facecolor=clrs[3])

ax1.set_title("Model1 Training Loss History").set_fontsize(15)
ax2.set_title("Model2 Training Loss History").set_fontsize(15)
ax1.set_ylabel("Loss").set_fontsize(15)
ax1.set_xlabel("Iterations").set_fontsize(15)
ax2.set_ylabel("Loss").set_fontsize(15)
ax2.set_xlabel("Iterations").set_fontsize(15)
ax1.legend()
ax2.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import genfromtxt
import itertools
```

## 1. Plot Training Loss History

```python
Model_Amplitude = []

for i in range(5):
    Model_Amplitude_PATH = f'/Users/tak/Github/QEmbedding/Results/earlystop 10 experiments/experiment{i+1}/Model Amplitude/Training_Loss_History.txt'
    Model_Amplitude.append(genfromtxt(Model_Amplitude_PATH, delimiter=',').tolist())

Model_Amplitude = list(itertools.zip_longest(*Model_Amplitude, fillvalue=np.nan))
Model_Amplitude_mean = np.array([np.nanmean(i) for i in Model_Amplitude])
Model_Amplitude_std = np.array([np.nanstd(i) for i in Model_Amplitude])


plt.rcParams['figure.figsize'] = [10, 5]
fig, ax = plt.subplots()
clrs = sns.color_palette("husl", 4)
with sns.axes_style("darkgrid"):
    ax.plot(range(len(Model_Amplitude_mean)), Model_Amplitude_mean, label="Model_Amplitude", c=clrs[0])
    ax.fill_between(range(len(Model_Amplitude_mean)), Model_Amplitude_mean-Model_Amplitude_std, Model_Amplitude_mean+Model_Amplitude_std, alpha=0.3,facecolor=clrs[0])

ax.set_title("Model Amplitude Training Loss History").set_fontsize(15)
ax.set_ylabel("Loss").set_fontsize(15)
ax.set_xlabel("Iterations").set_fontsize(15)
ax.legend()
```

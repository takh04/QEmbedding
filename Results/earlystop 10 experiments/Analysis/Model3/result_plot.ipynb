```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import genfromtxt
import itertools
```

## 1. Plot Training Loss History

```python
Model3_Fidelity, Model3_HSinner = [], []

for i in range(5):
    Model3_Fidelity_PATH = f'/Users/tak/Github/QEmbedding/Results/earlystop 10 experiments/experiment{i+1}/Model3 Fidelity/Training_Loss_History.txt'
    Model3_HSinner_PATH = f'/Users/tak/Github/QEmbedding/Results/earlystop 10 experiments/experiment{i+1}/Model3 HSinner/Training_Loss_History.txt'

    Model3_Fidelity.append(genfromtxt(Model3_Fidelity_PATH, delimiter=',').tolist())
    Model3_HSinner.append(genfromtxt(Model3_HSinner_PATH, delimiter=',').tolist())


Model3_Fidelity = list(itertools.zip_longest(*Model3_Fidelity, fillvalue=np.nan))
Model3_HSinner = list(itertools.zip_longest(*Model3_HSinner, fillvalue=np.nan))

Model3_Fidelity_mean = np.array([np.nanmean(i) for i in Model3_Fidelity])
Model3_Fidelity_std = np.array([np.nanstd(i) for i in Model3_Fidelity])

Model3_HSinner_mean = np.array([np.nanmean(i) for i in Model3_HSinner])
Model3_HSinner_std = np.array([np.nanstd(i) for i in Model3_HSinner])
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
fig, ax = plt.subplots()
clrs = sns.color_palette("husl", 4)
with sns.axes_style("darkgrid"):
    ax.plot(range(len(Model3_Fidelity_mean)), Model3_Fidelity_mean, label="Model3_Fidelity", c=clrs[0])
    ax.fill_between(range(len(Model3_Fidelity_mean)), Model3_Fidelity_mean-Model3_Fidelity_std, Model3_Fidelity_mean+Model3_Fidelity_std, alpha=0.3,facecolor=clrs[0])

    ax.plot(range(len(Model3_HSinner_mean)), Model3_HSinner_mean, label="Model3_HSinner", c=clrs[1])
    ax.fill_between(range(len(Model3_HSinner_mean)), Model3_HSinner_mean-Model3_HSinner_std, Model3_HSinner_mean+Model3_HSinner_std, alpha=0.3,facecolor=clrs[1])

ax.set_title("Model3 Training Loss History").set_fontsize(15)
ax.set_ylabel("Loss").set_fontsize(15)
ax.set_xlabel("Iterations").set_fontsize(15)
ax.legend()

```
